#### Read in 1000 rows of data

Each row is a move from a game (game_id), made by a player (player_id), we also have elo, white/black, etc.
"board" gives the "fen" - this the location of each piece on the position - prior to the move
"clock" gives the clock time at the end of the turn.
"time_control" specifies what the initial clock setting is for each player (e.g. 600 seconds) and also how much time they get back after each move (this comes after the "+"). "move" specifies what move the player made. "white_active" specifies which player (black or white) made the move

```python
data_raw = pd.read_csv(chess_data_fullfile, nrows=1000)
data_raw.head()
```

#### Filter data to only include moves from games with certain time-control settings, and compute estimated move-times for those games

```python
# what game types do we care about? filter dataframe to only include these time-control settings (these occur on first page) - new name for df is data_filt
game_time_types = ['60+0', '120+1', '180+0',
                   '180+2', '300+0', '300+3',
                   '600+0', '600+5', '900+10',
                   '1800+0', '1800+20']

# filter based on if it's in these game settings... 
data_filt = data_raw.loc[data_raw.time_control.isin(game_time_types)].reset_index(drop = True)

# remove first move of each player (this is the first two move_plys - half-moves - of the game)
def filter_moves(game_data):
    filt_game_data = game_data.loc[game_data.move_ply > 2].reset_index(drop = True)
    return filt_game_data

data_filt = data_filt.groupby('game_id').apply(filter_moves).reset_index(drop=True)

# compute rt for each move, add this to the dataframe (new name - data_rt)
def add_rt(game_df):
    
    game_df = game_df.reset_index(drop = True)
    game_time_setting = game_df.time_control[0]
    time_back = 0

    if '+1' in game_time_setting:
        time_back = 1
    elif '+2' in game_time_setting:
        time_back = 2
    elif '+3' in game_time_setting:
        time_back = 3
    elif '+5' in game_time_setting:
        time_back = 5
    elif '+10' in game_time_setting:
        time_back = 10
    elif '+20' in game_time_setting:
        time_back = 20
    
    white_move_rts = time_back + game_df.loc[game_df.white_active == True, 'clock'].diff()*-1
    black_move_rts = time_back + game_df.loc[game_df.white_active == False, 'clock'].diff()*-1
    game_df.loc[game_df.white_active == True, 'rt'] = white_move_rts
    game_df.loc[game_df.white_active == False, 'rt'] = black_move_rts
    
    return game_df

data_rt = data_filt.groupby('game_id').apply(add_rt).reset_index(drop = True)
```

#### How good was the move that the person made? - we want to evaluate this at multiple search depths

Afaik, we can't get stockfish to evaluate a specific move. Rather, we can only get it to evaluate a board state. Thus to evaluate a move, we push it onto the board and then evaluate the resultant board state. Stockfish evaluation proceeds w/ an iterative deepening search where at each step the depth is increased. Calling evaluations the way we do below prints out both the evaluation as well as the suggested line of moves (pv - principle variation) at the end of each iterative depth. To see this, set "print_everything" below to true. Here, we'll just grab the evaluation ('score') at the end of each depth and append it to a list. Once in a while (i think when the PV changes), stockfish repeats a depth. For simplicity, i'm just ignoring these cases. 

Note that we also need to clear the hash prior to each evaluation so that the depths have a consistent meaning.

```python
new_board = current_board.copy()
new_board.push_uci(this_move)
display(new_board)

max_depth = 15
last_depth = 0
score_over_depth = []

print_everything = True

# clear the hash
engine.configure({"Clear Hash": None})

with engine.analysis(new_board, multipv=1) as analysis:
    
    for info in analysis:
        
        if print_everything:
            print(info)
        
        this_depth = info.get("depth")
        
        # Handle re-searches
        if this_depth != last_depth:
        
            if info.get("pv") != None:
                score_over_depth.append(info.get('score',0))

                # Arbitrary stop condition.
                if (info.get('depth',0) >= max_depth):
                    break
                    
        last_depth = this_depth
        
print("Score over depth: ")
print(score_over_depth)
```

The score is at first framed in terms of the player whose move it is for that board position. Because we pushed the last move onto the board, it is now framed in terms of the opponents move. So we need to change the sign to get to corresopnd to the active player. Additionally, the score is in units of centipawns. Stockfish gives scores either in terms of centipawn advantage (1/100 of a pawn) or alternatively in distance to a mate. These need to be placed on a common scale. In our last paper, we estimated a logistic regression function to do this where we estimated win probability from either measure.

```python
# for cp to wp
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def cp_to_wp(score_val_cp_white):
    """
    Converts centipawn score val (signed with white pos/black neg) to win prob for white
    
    Args:
        score_val_cp_white: centipawn value (signed with white positive black negative) for position
    Returns:
        win probability: measure between 0 and 1 giving probability of white win
        Note: taken for logistic regression model applied to predict win given centipawn games
    """
    this_const = -0.045821224987387915;
    this_board_score_val = 0.002226678310106879
    #return sigmoid(log_reg_cp.params['const'] + log_reg_cp.params['board_score_val']*score_val_cp_white)
    return sigmoid(this_const + this_board_score_val*score_val_cp_white)

def mate_to_wp(score_val_mate):
    """
    Converts plys from mate to win prob (for the player who is within reach of mate)
    """
    # return sigmoid(log_reg_mate.params['const'] + log_reg_mate.params['score_val_pos']*np.abs(score_val_mate))
    return sigmoid(3.6986223572286208 + -0.05930670977061789*np.abs(score_val_mate))


# add support for if mate == 0!!!!!!!
def process_score(this_score, white_active):
    """
    Input:
    this score:  what is returned from stockfish
    white active: is whether it is white's move
    
    Output:
    score_dict: contains
    """
    
    white_score = this_score.white()
    is_mate = white_score.is_mate()
    
    if is_mate:
        score_type = 'mate'
        score_val = white_score.mate() # this is framed as white'
        if score_val > 0:
            wp = mate_to_wp(score_val) # this is framed as whichever side is close to mate
        elif score_val < 0:
            wp = 1 - mate_to_wp(score_val)
        elif score_val == 0:
            wp = 1 if white_active else 0
            # 1 or -1 depending on whose move this is.... so process score needs to take in who is active
    else: # centipawn...
        score_type = 'cp'
        score_val = white_score.score()
        wp = cp_to_wp(score_val)
    
    wp_active = wp if white_active else 1 - wp
        
    #score_dict = {'type': score_type, 'val': score_val, 'wp': wp}
    return wp_active

post_move_score_active_over_depth = np.array([process_score(this_score, this_white_active) for this_score in score_over_depth])
post_move_score_active_over_depth
```

The valuation on it's own tells us the board state valuation following the move, but doesn't really say how good the move was (we need to correct for how good the board state was prior to the move). I think we can just get the evaluation of the current state at multiple depths (this will also figure out a good move for each of those depths) and use this to subtract from the chosen move.

```python
max_depth = 15
last_depth = 0
score_over_depth = []

print_everything = False

# clear the hash
engine.configure({"Clear Hash": None})

with engine.analysis(current_board, multipv=1) as analysis:
    
    for info in analysis:
        
        if print_everything:
            print(info)
        
        this_depth = info.get("depth")
        
        # Handle re-searches
        if this_depth != last_depth:
        
            if info.get("pv") != None:
                score_over_depth.append(info.get('score',0))

                # Arbitrary stop condition.
                if (info.get('depth',0) >= max_depth):
                    break
                    
        last_depth = this_depth
        
print("Score over depth: ")
print(score_over_depth)

pre_move_score_active_over_depth = np.array([process_score(this_score, this_white_active) for this_score in score_over_depth])
pre_move_score_active_over_depth
```

```python
move_score_active_over_depth_corrected = post_move_score_active_over_depth - pre_move_score_active_over_depth
move_score_active_over_depth_corrected
```

In principle these should all be negative (because stockfish should be finding the best move), but in practice this is slightly complicated, becuase the evaluations are occuring 1 move-ply apart... i think this is fine, but we could think of alternative schemes for this.

Another route might be to try to find the moves which maximize the move utility metric (e.g. push the move onto the board and then evaluate at depth-n) that we're actually computing. We could do this by getting a consideration set of move suggestions, pushing each onto the board, evaluating them each at multiple depths and then taking the max score at each depth.

Here's how we might generate a consideration set of moves: 

#### Generate a large consideration set of possible moves

We want to find a list of moves that migth maximize the utility at each depth in the resultant board state

```python
# clear the hash
engine.configure({"Clear Hash": None})

#last_depth = 0

move_consideration_set = set()
depth_list = []
move_list = []

# we'll grab what stockfish thinks are the n best moves at each successive depth and append them to a set
n = 2

with engine.analysis(current_board, multipv=n) as analysis:
    for info in analysis:
        
        if info.get("pv") != None:
            
            this_move = info.get("pv")[0].uci()
            this_depth = info.get("depth",0)
            move_consideration_set.add(this_move)
            depth_list.append(this_depth)
            move_list.append(this_move)

            # Arbitrary stop condition.
            if (info.get('depth',0) >= 15):
                break

move_consideration_set
```

for each move in the consideration set, we'd then want to push it onto the board, and evaluate it at each successive depth from 1 to n.
